```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_clipboard()
df.head()
```

```python
sns.set(font_scale=1.2)
sns.set_style('whitegrid')
colors = sns.color_palette()
colors[0]
```

```python
def countplot(coltype, hue=None, yticks=range(0,26,5), order=None):
    if hue == None:
        cnts = df.value_counts(coltype)
        ax = sns.countplot(df, x=coltype, color=colors[0], order=order)
    else:
        cnts = df.value_counts([coltype, hue])
        ax = sns.countplot(df, x=coltype, hue=hue, order=order)
    plt.ylabel('Liczebność, n')
    if hue != None:
        shift = (ax.patches[1].get_width()) * 0.35
    else: 
        shift = 0.375
    for i, p in enumerate(ax.patches):
        ax.annotate(f'{p.get_height():2.0f}', (p.get_x()+shift, p.get_height()+0.15))
    plt.yticks(yticks)
    plt.show()

countplot('Płeć',)
countplot('Wiek', 'Płeć', yticks=range(0,17,2), order=sorted(df['Wiek'].unique()))
countplot('Stan cywilny', 'Płeć', yticks=range(0,21,2))
countplot('Wykształcenie', 'Płeć', yticks=range(0,17,2), order=['podstawowe', 'średnie/zawodowe', 'wyższe'])
countplot('Miejsce zamieszkania', 'Płeć', yticks=range(0,17,2))
countplot('Aktywny zawodowo', 'Płeć', yticks=range(0,17,2))
countplot('Czas leczenia (miesiące)', 'Płeć', yticks=range(0,11,2), order=sorted(df['Czas leczenia (miesiące)'].unique()))
```

```python
def countplot(coltype, hue=None, yticks=range(0,26,5), xticks = None, order=None):
    ax = sns.countplot(df, x=coltype, hue=hue)
    plt.xticks(xticks, order)
    plt.ylabel('Liczebność, n')
    shift = (ax.patches[1].get_width()) * 0.25
    for i, p in enumerate(ax.patches):
        ax.annotate(f'{p.get_height():2.0f}', (p.get_x()+shift, p.get_height()+0.15))
    plt.yticks(yticks)
    plt.xlabel('')
    plt.show()
```

```python
order = ['Nie wystąpił', 'Wcale', 'Trochę', 'Znacznie', 'Bardzo']

for i in range(1,9):
    countplot(f'P{i}', 'Płeć', range(0,17,2), xticks = range(5), order=order)
```
